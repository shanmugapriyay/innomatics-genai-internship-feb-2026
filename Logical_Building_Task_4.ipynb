```python
# ==============================================
# Question 1: Mobile Recharge Validation System
# ==============================================

# Function to validate recharge amount
def validate_recharge():

    # List of valid recharge plans
    valid_plans = [199, 299, 399, 599]

    # While loop to allow user retry until valid input
    while True:
        try:
            # Taking recharge amount from user
            amount = int(input("Enter recharge amount: ‚Çπ"))

            # Condition 1: Minimum recharge amount check
            if amount < 50:
                print("‚ùå Recharge amount must be at least ‚Çπ50")

            # Condition 2: Check if amount is in valid plans
            elif amount not in valid_plans:
                print("‚ùå Invalid plan. Available plans are:", valid_plans)

            # If all conditions are satisfied
            else:
                print("‚úÖ Recharge Successful!")
                break   # Exit loop when recharge is valid

        # Handling invalid input (non-numeric)
        except ValueError:
            print("‚ùå Please enter a valid numeric amount.")

# Function call
validate_recharge()
```

```python
# ==============================================
# Question 2: Inventory Reorder Alert System
# ==============================================

# Function to check stock levels
def inventory_alert(products):

    # Loop through dictionary items
    for product, stock in products.items():

        # If stock is less than 15 ‚Üí Reorder Alert
        if stock < 15:
            print(f"{product} ‚Üí üî¥ Reorder Alert (Stock: {stock})")

        # Otherwise stock is sufficient
        else:
            print(f"{product} ‚Üí üü¢ Stock OK (Stock: {stock})")


# Example product dictionary
inventory = {
    "Rice": 20,
    "Oil": 10,
    "Sugar": 25,
    "Salt": 8
}

# Function call
inventory_alert(inventory)
```

```python
# ==============================================
# Question 3: Student Result Processing System
# ==============================================

# Function to calculate average and result
def student_result(marks):

    total = 0  # Variable to store total marks

    # Loop to calculate total marks
    for mark in marks:
        total += mark

    # Calculate average
    average = total / len(marks)
    print("Average Marks:", average)

    # Check pass or fail condition
    if average >= 50:
        return "‚úÖ Pass"
    else:
        return "‚ùå Fail"


# Example marks list
marks_list = [60, 45, 70, 50]

# Function call and storing result
result = student_result(marks_list)

# Print final result
print("Result:", result)
```

```python
# ==============================================
# Question 4: Cab Fare Estimator with Retry Option
# ==============================================

# Function to calculate fare
def calculate_fare(distance, peak):

    base_fare = 50            # Fixed base fare
    fare = base_fare + (12 * distance)  # Distance charge

    # If peak hour, add 25% extra
    if peak.lower() == "yes":
        fare += fare * 0.25

    return fare


# While loop to allow retry
while True:
    try:
        # Taking user input
        distance = float(input("Enter distance in km: "))
        peak = input("Is it peak hour? (yes/no): ")

        # Function call
        total_fare = calculate_fare(distance, peak)

        # Display fare
        print(f"üöñ Total Fare: ‚Çπ{total_fare}")

        # Asking user for retry
        retry = input("Do you want to calculate again? (yes/no): ")

        # Exit loop if user says no
        if retry.lower() != "yes":
            break

    except ValueError:
        print("‚ùå Please enter valid numeric distance.")
```

```python
# ==============================================
# Question 5: Employee Attendance Eligibility Checker
# ==============================================

# Function to calculate attendance eligibility
def check_eligibility(attendance_list):

    present_days = 0  # Counter for present days

    # Loop to count "P"
    for day in attendance_list:
        if day == "P":
            present_days += 1

    # Calculate attendance percentage
    percentage = (present_days / len(attendance_list)) * 100
    print("Attendance Percentage:", percentage)

    # Check eligibility condition
    if percentage >= 75:
        return "‚úÖ Eligible"
    else:
        return "‚ùå Not Eligible"


# Example attendance list
attendance = ["P", "P", "A", "P", "P", "A", "P", "P"]

# Function call
status = check_eligibility(attendance)

# Print result
print("Status:", status)
```

```python
# ==============================================
# Question 6: Password Strength Checker
# ==============================================

# Function to check password strength
def check_password(password):

    has_digit = False       # Flag to check digit
    has_special = False     # Flag to check special character
    special_chars = "@#$"   # Allowed special characters

    # Condition 1: Minimum length check
    if len(password) < 8:
        return "‚ùå Password must be at least 8 characters long"

    # Loop through each character in password
    for char in password:

        # Check if character is a digit
        if char.isdigit():
            has_digit = True

        # Check if character is special
        if char in special_chars:
            has_special = True

    # Final validation
    if has_digit and has_special:
        return "‚úÖ Strong Password"
    else:
        return "‚ùå Weak Password (Must contain at least one digit and one @#$ character)"


# Taking password input from user
user_password = input("Enter password: ")

# Function call and print result
print(check_password(user_password))
```
