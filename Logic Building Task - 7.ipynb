```python
# ----------------------------------------
# Question 1: Smart Light Controller
# ----------------------------------------

# Create a class for Smart Light
class SmartLight:
    
    # Constructor to initialize light name
    def __init__(self, light_name):
        self.light_name = light_name
        self.status = "OFF"   # Default status is OFF
    
    # Method to turn ON the light
    def turn_on(self):
        self.status = "ON"
    
    # Method to turn OFF the light
    def turn_off(self):
        self.status = "OFF"
    
    # Method to display light status
    def display_status(self):
        print(f"{self.light_name} is {self.status}")


# Sample Usage
light = SmartLight("Bedroom Light")
light.turn_on()
light.display_status()
```

```python
# ----------------------------------------
# Question 2: Employee ID Card System
# ----------------------------------------

# Create a class for Employee ID Card
class Employee:
    
    # Constructor to initialize employee details
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department
    
    # Method to display ID card details
    def display_id_card(self):
        print("Employee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.emp_id}")
        print(f"Department: {self.department}")


# Sample Usage
emp = Employee("Rahul", "EMP102", "IT")
emp.display_id_card()
```

```python
# ----------------------------------------
# Question 3: Mobile Contact Record
# ----------------------------------------

# Create a class for Contact
class Contact:
    
    # Constructor to initialize contact details
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone
    
    # Method to display contact information
    def display_contact(self):
        print("Contact Saved")
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone}")


# Sample Usage
contact = Contact("Anita", "9876543210")
contact.display_contact()
```

```python
# ----------------------------------------
# Question 4: Product Price Tag Generator
# ----------------------------------------

# Create a class for Product
class Product:
    
    # Constructor to initialize product details
    def __init__(self, product_name, price):
        self.product_name = product_name
        self.price = price
    
    # Method to display price tag
    def display_price_tag(self):
        print(f"Product: {self.product_name}")
        print(f"Price: ₹{self.price}")


# Sample Usage
product = Product("Headphones", 2499)
product.display_price_tag()
```

```python
# ----------------------------------------
# Question 5: Movie Rating Display System
# ----------------------------------------

# Create a class for Movie
class Movie:
    
    # Constructor to initialize movie details
    def __init__(self, movie_name, rating):
        self.movie_name = movie_name
        self.rating = rating
    
    # Method to display movie details
    def display_movie(self):
        print(f"Movie: {self.movie_name}")
        print(f"Rating: {self.rating} / 5")


# Sample Usage
movie = Movie("Inception", 4.8)
movie.display_movie()
```

```python
# ----------------------------------------
# Question 6: Delivery Address Manager
# ----------------------------------------

# Create a class for Delivery Details
class Delivery:
    
    # Constructor to initialize customer details
    def __init__(self, customer_name, address):
        self.customer_name = customer_name
        self.address = address
    
    # Method to display delivery details
    def display_delivery(self):
        print("Delivery Details")
        print(f"Customer: {self.customer_name}")
        print(f"Address: {self.address}")


# Sample Usage
delivery = Delivery("Suman", "Hyderabad")
delivery.display_delivery()
```
