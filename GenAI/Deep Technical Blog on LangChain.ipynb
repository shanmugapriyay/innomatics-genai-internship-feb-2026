Import Libraries

```python
import os

from langchain_groq import ChatGroq
from langchain_core.prompts import PromptTemplate
from langchain.chains import LLMChain
from langchain.memory import ConversationBufferMemory
from langchain.chains import ConversationChain
from langchain.agents import initialize_agent, load_tools
```

Add groq api

```python
os.environ["GROQ_API_KEY"] = "Give your api key"
```

Initialize the LLM

```python
llm = ChatGroq(
    model="llama-3.3-70b-versatile",
    temperature=0.7
)

response = llm.invoke("Explain LangChain in simple terms")

print(response.content)
```

Prompt Template Example

```python
template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple beginner-friendly terms."
)

prompt = template.format(topic="LangChain")

print(llm.invoke(prompt).content)
```

LLMChain Example

```python
chain = LLMChain(
    llm=llm,
    prompt=template
)

result = chain.run("AI Agents")

print(result)
```

Memory Example

```python
memory = ConversationBufferMemory()

conversation = ConversationChain(
    llm=llm,
    memory=memory
)

print(conversation.predict(input="Hi my name is Gayathri"))
print(conversation.predict(input="What is my name?"))
```

Agent Example (Tool Use)

```python
tools = load_tools(["llm-math"], llm=llm)

agent = initialize_agent(
    tools,
    llm,
    agent="zero-shot-react-description",
    verbose=True
)

agent.run("What is 25 * 17?")
```

Document Loader Example

```python
from langchain.document_loaders import TextLoader

loader = TextLoader("Port.txt")

documents = loader.load()

print(documents)
```
