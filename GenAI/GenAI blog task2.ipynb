## Section 1: Introduction to LangChain
LangChain is an open-source framework that simplifies building applications using Large Language Models (LLMs). It acts like a bridge between language models and real-world applications such as chatbots, QA systems, document search tools, and autonomous AI agents.

## Section 2: Core Components of LangChain

### 2.1 LLMs and Chat Models

```python
from langchain_groq import ChatGroq
import os
from google.colab import userdata

# Ensure GROQ_API_KEY is stored in Colab Secrets
# To add it: click the '🔑' icon in the left panel, then '+ New secret',
# enter 'GROQ_API_KEY' for Name and your key for Value.

GROQ_API_KEY = userdata.get('GROQ_API_KEY')
os.environ["GROQ_API_KEY"] = GROQ_API_KEY

llm = ChatGroq(
    model="llama-3.3-70b-versatile",
    temperature=0.7
)

response = llm.invoke("Explain LangChain in simple terms")
print(response.content)
```

LLMs are the brain of LangChain. They generate intelligent responses based on user input.

### 2.2 Prompt Templates

```python
from langchain_core.prompts import PromptTemplate

prompt = PromptTemplate.from_template(
    "Explain {topic} in simple words for beginners."
)

formatted_prompt = prompt.format(topic="Machine Learning")
print(formatted_prompt)
```

Prompt templates help create reusable dynamic prompts.

### 2.3 Chains

```python
# Using LangChain Expression Language (LCEL) for chaining prompt and LLM
chain = prompt | llm

# For LCEL chains, use .invoke() with a dictionary matching prompt variables
result = chain.invoke({"topic": "Artificial Intelligence"})
print(result.content)
```

Chains connect prompts and LLMs into workflows.

Memory stores previous conversation context.

```python
from langchain.memory import ConversationBufferMemory
from langchain_community.chat_message_histories import ChatMessageHistory

# Create chat history
chat_history = ChatMessageHistory()

# Create memory object
memory = ConversationBufferMemory(
    chat_memory=chat_history,
    return_messages=True
)

# Save conversation
memory.save_context(
    {"input": "My name is Priya"},
    {"output": "Nice to meet you Priya"}
)

memory.save_context(
    {"input": "I love Python"},
    {"output": "Python is great for AI development"}
)

# Display stored memory
print(memory.load_memory_variables({}))
```

### 2.5 Tools and 2.6 Agents

```python
from langchain.tools import tool
from langchain.agents import initialize_agent

@tool
def calculator(query: str):
    """Simple calculator tool"""
    return eval(query)

tools = [calculator]

agent = initialize_agent(
    tools,
    llm,
    agent="zero-shot-react-description",
    verbose=True
)

agent_result = agent.run("What is 25 multiplied by 8?")
print(agent_result)
```

Agents decide automatically which tool to use.

### 2.7 Document Loaders

```python
from langchain_community.document_loaders import TextLoader

with open("sample.txt", "w") as file:
    file.write("LangChain helps developers build AI applications.")

loader = TextLoader("sample.txt")
documents = loader.load()
print(documents)
```

Document loaders import external files into LangChain.

### 2.8 Vector Stores

```python
from langchain_community.vectorstores import FAISS
from langchain_huggingface import HuggingFaceEmbeddings

texts = [
    "LangChain is useful for AI applications.",
    "FAISS is used for vector search.",
    "Embeddings convert text into vectors."
]

embedding_model = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

vectorstore = FAISS.from_texts(texts, embedding_model)
results = vectorstore.similarity_search("What is vector search?")

for doc in results:
    print(doc.page_content)
```

Vector stores enable semantic search using embeddings.
