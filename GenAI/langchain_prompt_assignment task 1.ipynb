```python
from langchain_core.prompts import PromptTemplate
from langchain_core.prompts import ChatPromptTemplate, SystemMessagePromptTemplate, HumanMessagePromptTemplate
```

## Task 1: Replace Hardcoded Prompts

```python
template_1 = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} in simple terms for beginners"
)

print(template_1.format(topic="Artificial Intelligence"))
```

## Task 2: Multi-Input Prompt System

```python
template_2 = PromptTemplate(
    input_variables=["topic", "audience", "tone"],
    template="Explain {topic} for {audience} in a {tone} tone"
)

test_cases = [
    {"topic": "AI", "audience": "beginners", "tone": "friendly"},
    {"topic": "Python", "audience": "kids", "tone": "fun"},
    {"topic": "Deep Learning", "audience": "engineers", "tone": "technical"}
]

for case in test_cases:
    print(template_2.format(**case))
```

## Task 3: Prompt Variations Engine

```python
teaching_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} clearly step by step"
)

interview_template = PromptTemplate(
    input_variables=["topic"],
    template="Ask 3 questions about {topic}"
)

story_template = PromptTemplate(
    input_variables=["topic"],
    template="Explain {topic} as a story"
)

topic = "Machine Learning"

print(teaching_template.format(topic=topic))
print(interview_template.format(topic=topic))
print(story_template.format(topic=topic))
```

## Task 4: ChatPromptTemplate System

```python
def get_chat_prompt(role, topic):
    if role == "teacher":
        system_template = "You are a helpful teacher."
    elif role == "interviewer":
        system_template = "You are an interviewer."
    elif role == "motivator":
        system_template = "You are a motivator."
    else:
        system_template = "You are a helpful assistant."

    system_message = SystemMessagePromptTemplate.from_template(system_template)
    human_message = HumanMessagePromptTemplate.from_template("Talk about {topic}")

    chat_prompt = ChatPromptTemplate.from_messages([system_message, human_message])
    return chat_prompt.format_messages(topic=topic)

print(get_chat_prompt("teacher", "Neural Networks"))
```

## Task 5: Input Validation

```python
def validate_inputs(audience, tone):
    valid_audience = ["beginner", "intermediate", "expert"]
    valid_tone = ["formal", "casual", "fun"]

    if audience not in valid_audience:
        audience = "beginner"

    if tone not in valid_tone:
        tone = "formal"

    return audience, tone

print(validate_inputs("kids", "friendly"))
```

## Task 6: Prompt Generator App

```python
def generate_prompt(topic, audience, tone, style):
    audience, tone = validate_inputs(audience, tone)

    templates = {
        "teaching": "Explain {topic} for {audience} in a {tone} teaching style",
        "interview": "Ask interview questions about {topic}",
        "storytelling": "Explain {topic} in a storytelling way"
    }

    selected_template = templates.get(style)

    prompt = PromptTemplate(
        input_variables=["topic", "audience", "tone"],
        template=selected_template
    )

    return prompt.format(topic=topic, audience=audience, tone=tone)

print(generate_prompt("Neural Networks", "beginner", "fun", "storytelling"))
```
