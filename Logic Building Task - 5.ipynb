```python
# Problem 1: Social Media – Post Engagement Analyzer

def analyze_post_engagement(likes_list):
    # Calculate total likes using loop
    total_likes = 0
    for like in likes_list:
        total_likes += like

    # Check if post is viral
    if total_likes >= 1000:
        status = "Viral Post"
    else:
        status = "Normal Engagement"

    # Display results
    print("Total Likes:", total_likes)
    print("Post Status:", status)


# Sample Data
likes = [200, 300, 150, 400]
analyze_post_engagement(likes)
```

```python
# Problem 2: Medicine Stock Alert System

def check_medicine_stock(stock):
    # Check stock condition
    if stock < 10:
        status = "Low Stock Alert"
    else:
        status = "Stock Sufficient"

    print("Medicine Stock:", stock)
    print("Status:", status)


# Sample Input
check_medicine_stock(6)
```

```python
# Problem 3: Rainfall Adequacy Checker

def check_rainfall(rainfall_list, required_level):
    total = 0

    # Calculate total rainfall
    for rain in rainfall_list:
        total += rain

    # Calculate average rainfall
    average = total / len(rainfall_list)

    # Check adequacy
    if average >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    print("Average Rainfall:", average)
    print("Rainfall Status:", status)


# Sample Data
rainfall_data = [60, 80, 70, 78]
check_rainfall(rainfall_data, 70)
```

```python
# Problem 4: Duplicate Account Detection

def check_duplicate_accounts(usernames):
    # Convert list to set
    unique_users = set(usernames)

    # Compare lengths
    if len(unique_users) != len(usernames):
        result = "Yes"
    else:
        result = "No"

    print("Duplicate Accounts Found:", result)


# Sample Data
users = ["john", "mary", "alex", "john"]
check_duplicate_accounts(users)
```

```python
# Problem 6: Premium Crop Price Filter

def filter_premium_crops(price_list):
    premium_crops = []

    # Filter prices above ₹2000
    for price in price_list:
        if price > 2000:
            premium_crops.append(price)

    print("Premium Crops:", premium_crops)


# Sample Data
crop_prices = [1500, 2500, 1800, 3200]
filter_premium_crops(crop_prices)
```

```python
# Problem 7: Application Health Checker

def check_application_health(error_count):
    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    print("Error Count:", error_count)
    print("System Status:", status)


# Sample Input
check_application_health(7)
```

```python
# Problem 8: Daily Transaction Limit Checker

def check_transaction(amount):
    daily_limit = 50000

    if amount <= daily_limit:
        status = "Approved"
    else:
        status = "Rejected"

    print("Transaction Amount:", amount)
    print("Transaction Status:", status)


# Sample Input
check_transaction(60000)
```

```python
# Problem 9: Student Attendance Eligibility

def check_attendance(attendance_list):
    total_days = len(attendance_list)
    present_days = 0

    # Count present days
    for day in attendance_list:
        if day == "P":
            present_days += 1

    # Calculate percentage
    percentage = (present_days / total_days) * 100

    # Check eligibility
    if percentage >= 75:
        status = "Eligible"
    else:
        status = "Not Eligible"

    print("Attendance Percentage:", percentage)
    print("Exam Eligibility:", status)


# Sample Data
attendance = ["P", "P", "A", "P", "P"]
check_attendance(attendance)
```

```python
# Problem 10: Smart Electricity Bill Analyzer

def calculate_electricity_bill(units):
    bill = 0

    # Slab Calculation
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = (100 * 3) + ((units - 100) * 5)
    else:
        bill = (100 * 3) + (100 * 5) + ((units - 200) * 7)

    # Usage Classification
    if bill < 500:
        status = "Low Usage"
    elif 500 <= bill <= 1500:
        status = "Moderate Usage"
    else:
        status = "High Usage"

    return bill, status


# Sample Input
total_bill, usage_status = calculate_electricity_bill(250)

print("Total Bill:", total_bill)
print("Usage Level:", usage_status)
```
