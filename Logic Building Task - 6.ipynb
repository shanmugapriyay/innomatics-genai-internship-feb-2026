```python
# ---------------------------------------------------------
# Question 1: Smart Parking Lot Management System
# ---------------------------------------------------------

def smart_parking(capacity, logs):
    parked = 0          # Current parked vehicles
    peak = 0            # Highest parking count recorded

    # Loop through each vehicle log
    for log in logs:
        if log == "IN":
            parked += 1
        elif log == "OUT" and parked > 0:
            parked -= 1

        # Track peak usage
        if parked > peak:
            peak = parked

    print("Currently Parked Vehicles:", parked)

    # Capacity check
    if parked > capacity:
        print("Parking Status: Full - Capacity Exceeded!")
    else:
        print("Parking Status: Available")

    print("Peak Usage:", peak)


# Sample Input
smart_parking(50, ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"])
```

```python
# ---------------------------------------------------------
# Question 2: Online Food Delivery Time Estimator
# ---------------------------------------------------------

def estimate_delivery_time(distance, traffic, weather):
    
    # Step 1: Calculate base delivery time
    # Assuming 5 minutes per km
    base_time = distance * 5
    
    # Step 2: Add delay based on traffic condition
    if traffic == "High":
        base_time += 10
    elif traffic == "Medium":
        base_time += 5
    # Low traffic → No extra delay

    # Step 3: Add delay based on weather condition
    if weather == "Rainy":
        base_time += 5
    elif weather == "Storm":
        base_time += 15
    # Clear weather → No extra delay

    # Final ETA Output
    print("Estimated Delivery Time:", base_time, "minutes")


# ---------------------------------------------------------
# Sample Input
# ---------------------------------------------------------
estimate_delivery_time(8, "High", "Rainy")
```

```python
# ---------------------------------------------------------
# Question 3: Movie Theatre Seat Occupancy Analyzer
# ---------------------------------------------------------

def theatre_occupancy(total_seats, booked_seats):

    # Count booked seats
    booked = len(booked_seats)

    # Calculate occupancy percentage
    occupancy = (booked / total_seats) * 100

    print("Occupancy:", round(occupancy), "%")

    # Show status classification
    if occupancy == 100:
        print("Show Status: Housefull")
    elif occupancy >= 75:
        print("Show Status: Almost Full")
    else:
        print("Show Status: Seats Available")


# Sample Input
theatre_occupancy(200, [1] * 150)
```

```python
# ---------------------------------------------------------
# Question 4: Cloud Server Load Classification System
# ---------------------------------------------------------

def server_load(cpu_readings):

    # Calculate average CPU usage
    average = sum(cpu_readings) / len(cpu_readings)

    print("Average CPU Load:", round(average), "%")

    # Classification based on rules
    if average < 50:
        print("Server Status: Normal")
    elif 50 <= average <= 80:
        print("Server Status: Warning")
    else:
        print("Server Status: Critical")


# Sample Input
server_load([45, 60, 70, 85, 90])
```

```python
# ---------------------------------------------------------
# Question 5: Smart Classroom Resource Usage Monitor
# ---------------------------------------------------------

def classroom_monitor(resources):

    overused = []  # List to store overused resources

    # Loop through dictionary items
    for resource, hours in resources.items():
        if hours > 8:  # Threshold for overuse
            overused.append(resource)

    # Output result
    if overused:
        print("Overused Resources:", ", ".join(overused))
        print("Energy Alert: Yes")
    else:
        print("Overused Resources: None")
        print("Energy Alert: No")


# Sample Input
classroom_monitor({
    "Projector": 6,
    "AC": 9,
    "Lights": 4
})
```

```python
# ---------------------------------------------------------
# Question 6: Online Event Registration Capacity Controller
# ---------------------------------------------------------

def event_registration(capacity, registrations):

    # Check if registrations exceed capacity
    if registrations > capacity:
        confirmed = capacity
        waitlisted = registrations - capacity
        status = "Closed"
    else:
        confirmed = registrations
        waitlisted = 0
        status = "Open"

    print("Confirmed Registrations:", confirmed)
    print("Waitlisted Users:", waitlisted)
    print("Registration Status:", status)


# Sample Input
event_registration(100, 105)
```
