```python
# 1. USER LOGIN CHECK

# Predefined credentials (stored in system)
username = "admin"
password = "1234"

# Input credentials (simulate user input)
input_username = "admin"     # Change this to test
input_password = "1234"      # Change this to test

# Checking login credentials
if input_username == username and input_password == password:
    print("Login Successful")
else:
    print("Invalid Credentials")
```

```python
# 2. PASS / FAIL ANALYZER

# List of student marks
marks = [45, 78, 90, 33, 60]

# Initialize counters
pass_count = 0
fail_count = 0

# Analyze each mark
for mark in marks:
    if mark >= 50:
        pass_count += 1
    else:
        fail_count += 1

# Print final results
print("Total Students:", len(marks))
print("Number of Pass Students:", pass_count)
print("Number of Fail Students:", fail_count)
```

```python
# 3. SIMPLE DATA CLEANER

# Original list with messy data
names = [" Alice ", "bob", " CHARLIE "]

# Cleaned list
cleaned_names = []

# Cleaning process
for name in names:
    cleaned_name = name.strip().lower()  # Remove spaces + convert to lowercase
    cleaned_names.append(cleaned_name)

# Print cleaned result
print("Original Names:", names)
print("Cleaned Names:", cleaned_names)
```

```python
# 4. MESSAGE LENGTH ANALYZER

# List of messages
messages = ["Hi", "Welcome to the platform", "OK"]

# Analyze each message
for message in messages:
    length = len(message)
    print(f"Message: '{message}' | Length: {length}")
    
    # Flag messages longer than 10 characters
    if length > 10:
        print("This message is longer than 10 characters.")
    
    print("-" * 40)
```

```python
# 5. ERROR MESSAGE DETECTOR

# System logs
logs = ["INFO", "ERROR", "WARNING", "ERROR"]

# Count ERROR entries
error_count = logs.count("ERROR")

# Print result
print("System Logs:", logs)
print("Total ERROR Entries:", error_count)
```
